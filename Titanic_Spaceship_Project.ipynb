```python
#Creating function to take care of Null values
def split_cabin(x):
  if len(str(x).split('/'))<3:
     return['Missing', 'Missing' ,'Missing']
  else:
     return str(x).split('/')


```

```python
#create function for data tranformation
def preprocessing(df_train):
  ##Fill the missing values in Homeplanet as 'Missing'
  df_train["HomePlanet"].fillna('Missing',inplace=True)
  #Seperating Deck and Side fro Cabin
  df_train['TempCabin']=df_train['Cabin'].apply(lambda x: split_cabin(x))
  df_train['Deck']=df_train['TempCabin'].apply(lambda x: x[0])
  df_train['Side']=df_train['TempCabin'].apply(lambda x: x[2])
  # #CryoSleep fill with 'Missing'
  df_train['CryoSleep'].fillna('Missing',inplace=True)
  # #Destination fill with 'Missing'
  df_train['Destination'].fillna('Missing',inplace= True)
  # #Age- fill with mean
  df_train['Age'].fillna(df_train['Age'].mean(),inplace= True)
  #  #VIP drop nulls
  df_train["VIP"].fillna('Missing',inplace=True)
  # #Name-Dropping name column due to high cardinality
  df_train.drop(columns=['Name'], inplace=True)
  # #Other spending Columns 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'
  df_train['RoomService'].fillna(0,inplace= True)
  df_train[ 'FoodCourt'].fillna(0,inplace= True)
  df_train['ShoppingMall'].fillna(0,inplace= True)
  df_train['Spa'].fillna(0,inplace= True)
  df_train['VRDeck'].fillna(0,inplace= True)
  # # df_train.dropna(inplace=True)
  df_train.drop(['TempCabin','Cabin'],axis=1,inplace=True)

preprocessing(df_train)




```

```python
#Modeling. Creating target & Features, X&Y value, one hot encoding, Train and split, train on algorithmns
```

```python

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python

X =abt.drop(["Transported",'PassengerId'],axis=1)
X= pd.get_dummies(X)
y =abt["Transported"]
```

```python
#Spliting datasets into two sets for training and testing.
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=1234)


```

```python
#Setting up ML pipelines

Pipelines = { "rf": make_pipeline(StandardScaler(),RandomForestClassifier(random_state=1234)),
             "gb": make_pipeline(StandardScaler(),GradientBoostingClassifier(random_state=1234))

}
grid ={ "rf" :{ "randomforestclassifier__n_estimators":[100,200,300]},
    "gb" :{"gradientboostingclassifier__n_estimators":[100,200,300]}

}

```

```python
Pipelines.items()
```

```python
#Creating an emply dictionary ti fit the models
fit_models={}
```

```python
X_train
```

```python
#looping through the algorithms
for algo,pipeline in Pipelines.items():
  #Create new gridsearchcv class
  model = GridSearchCV(pipeline,grid[algo],n_jobs=-1,cv=10)
  #fit into models
  model.fit(X_train, y_train)
  #Store the results into the dictionary
  fit_models[algo] = model
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
for algo, model in fit_models.items():
  yhat= model.predict(X_test)
  accuracy = accuracy_score(y_test,yhat)
  precision = precision_score(y_test,yhat)
  recall = recall_score(y_test,yhat)
  print(f'Metrics for {algo}: accuracy-{accuracy},recall:{recall},precision:{precision}')
```

```python
df_test = pd.read_csv('/content/test.csv')
```

```python
#making a copy of test_csv
abt_test = df_test.copy()
#passing through preprocessing function
preprocessing(abt_test)
#one hot encoding
abt_test = pd.get_dummies(abt_test.drop('PassengerId',axis=1))
```

```python
final_test = fit_models['gb'].predict(abt_test)
submission = pd.DataFrame([df_test['PassengerId'],final_test]).T
submission.columns=['PassengerId','Transported']
```
